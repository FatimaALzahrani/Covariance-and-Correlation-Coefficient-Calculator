# **Covariance and Correlation Calculator üìà**
Welcome to the interactive tool for calculating covariance and correlation coefficient!  
You can generate data, calculate statistics, and visualize the relationship between variables.  
Follow the steps below to explore the relationships between X and Y.

##üìã **Generate Data Function**

This function creates two variables, X and Y, with a linear relationship and some noise.

```python
import numpy as np

def generate_data(seed=42, n=100, noise=0.5, slope=2):
    np.random.seed(seed)
    x = np.random.randn(n)
    y = slope * x + np.random.randn(n) * noise
    return x, y
```

## **üìä Calculate Statistics**

This function computes covariance, standard deviations, and the correlation coefficient.

```python
def calculate_statistics(x, y):
    mean_x, mean_y = np.mean(x), np.mean(y)
    covariance = np.sum((x - mean_x) * (y - mean_y)) / (len(x) - 1)
    std_x, std_y = np.std(x, ddof=1), np.std(y, ddof=1)
    correlation = covariance / (std_x * std_y)
    return {
        "mean_x": mean_x,
        "mean_y": mean_y,
        "covariance": covariance,
        "std_x": std_x,
        "std_y": std_y,
        "correlation": correlation,
    }
```

## **üé® Scatter Plot Function**

This function creates a scatter plot of X and Y with mean lines and correlation.

```python
import matplotlib.pyplot as plt

def plot_data(x, y, stats):
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, alpha=0.7, label="üìç Data Points")
    plt.axhline(stats['mean_y'], color="red", linestyle="--", label="üî¥ Mean Y")
    plt.axvline(stats['mean_x'], color="blue", linestyle="--", label="üîµ Mean X")
    plt.title(f"Scatter Plot üìä (Correlation: {stats['correlation']:.4f})")
    plt.xlabel("X-axis ‚û°Ô∏è")
    plt.ylabel("Y-axis ‚¨ÜÔ∏è")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

## **üåü Main Program**

This cell interacts with the user to generate data, calculate statistics, and visualize results.

```python
print("üåü Welcome to the Covariance and Correlation Calculator üåü")

n = int(input("üî¢ Enter the number of data points (e.g., 100): "))
noise = float(input("üéõÔ∏è Enter the noise level (e.g., 0.5): "))
slope = float(input("üìà Enter the slope of the linear relationship (e.g., 2): "))
seed = int(input("üé≤ Enter a random seed for reproducibility (e.g., 42): "))

x, y = generate_data(seed=seed, n=n, noise=noise, slope=slope)
stats = calculate_statistics(x, y)

print("\nüìã --Results:--")
print(f"   ‚û°Ô∏è Covariance: {stats['covariance']:.4f}")
print(f"   ‚û°Ô∏è Correlation Coefficient: {stats['correlation']:.4f}")

print("\nüîç Visualizing your data...")
plot_data(x, y, stats)
```
